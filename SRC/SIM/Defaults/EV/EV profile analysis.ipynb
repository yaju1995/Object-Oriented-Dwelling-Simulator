```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("pdf_Veh1_Level1.csv")
```

```python
df
```

```python

columns_to_plot = ["duration", "start_soc", "start_time"]

df[columns_to_plot].hist(bins=30, figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
df["start_time"].hist(bins=40)
plt.xlabel("`Plug in Time")
plt.ylabel("Count")
plt.title("Histogram")
plt.show()
```

```python
df_weekday = df[df["weekday"] == 1]
df_weekend = df[df["weekday"] == 0]
```

```python
df_weekday[columns_to_plot].hist(bins=30, figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
df["duration"].hist(bins=40)
plt.xlabel("Duration")
plt.ylabel("Count")
plt.title("Histogram of Duration")
plt.show()
```
